# Hyperparameter Exploration

Our learnings from changing the learning rate and weight decay values. We've typically been going with 1e-4 and 0.05, respectively, for sae-ception on our best ViT-H model; However, the optimal parameters here might be (1e-5, 0.05) or (1e-5, 0.1).

Though this is just for the resulting accuracy of the model... we'd need to do some runs to verify this.

```python
import os
import pickle

base_dir = './features/classifier-99.37/25_top/hyperparams-99.37'
all_loss_data = {}

for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            if file.endswith('.pkl'):
                file_path = os.path.join(folder_path, file)
                with open(file_path, 'rb') as f:
                    loss_data = pickle.load(f)
                all_loss_data[folder] = loss_data
```
